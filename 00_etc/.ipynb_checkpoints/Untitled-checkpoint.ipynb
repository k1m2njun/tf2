```python
import pandas as pd
```

```python
df_b = pd.read_csv('시군구_성_월별_출생_20220925213251.csv', encoding='euc-kr')
df_b
```

```python
df_b.rename(columns={'전국':'출생자(명)'}, inplace=True)
df_b = df_b.drop([0])
df_b
```

```python
df_b['출생자(명)'] = df_b['출생자(명)'].astype(int)
df_b.info()
```

```python
df_b.plot(x='시점',y='출생자(명)',kind='line')
```

```python
df_d = pd.read_csv('사망자수__조사망률__기대수명_20220925215132.csv', encoding='euc-kr')
df_d
```

```python
df_d.plot(x='시점',y='사망자수(명)',kind='line')
```

```python
import seaborn as sns
```

```python
df_bd = pd.merge(df_b, df_d, left_on='시점', right_on='시점', how='outer')
df_bd = df_bd[['시점', '출생자(명)', '사망자수(명)']]
df_bd.drop([24, 25], inplace=True) # 21년도 데이터 삭제
df_bd
```

```python
df_bd = df_bd.set_index('시점')
df_bd
```

```python
df_bd
```

```python
df_bd.plot()
```

```python
df_bd = df_bd.cumsum()
```

```python
import seaborn as sns
```

```python
sns.displot(data=df_bd)
```
