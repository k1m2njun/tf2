# JSon 을 이용한 영역표시
데이터 교환 형식. text 기반(파일을 열어서 직접 읽어 볼 수 있다.) 다양한 언어를 지원한다.  
파이썬에서는 json 파일을 받아서 리스트, 튜플 등의 데이터 형식으로 바꾸어 불러올 수 있다.  
다른 언어 사이에서 데이터를 서로 주고 받을 때 사용할 수 있는 문법이다.
csv는 2차원배열구조로 한계가 있지만, json은 더 복잡한 구조의 표현이 가능하다.

# GeoJSon
선, 다각형, 폴리곤 등 데이터를 표현할 수 있다.
예를 들어 여러군데의 지역 위경도를 입력해서 지도상 영역을 표시할 수 있다.

```python
jeju_loc = [33.3616819, 126.5291548]
```

```python
import json
```

```python
m = folium.Map(jeju_loc, zoom_start=11)
json_data = open('jejusi.json', 'r').read() # peakle 파일에서는 바이너리라서 rb 등 처럼 b를 붙인다.
json_data # 여기서 보면 str 형식의 데이터임을 알 수 있다.
```

```python
json_data1 = json.loads(json_data) # 용량이 클수록 오류가 날 수 있어서 한 번 읽어온 뒤에 로드하는 게 좋다.
json_data1 # 반면 데이터로 불러와서 보니 딕셔너리 형태의 데이터로 불러왔다.
```

```python
folium.GeoJson(json_data1).add_to(m)
m # geojson을 이용해 이렇게 영역을 표시할 수 있다.
```
