```python
# 맑은고딕 Malgun Gothic
# 폰트 세팅
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
matplotlib.get_configdir()
```

```python
matplotlib.get_cachedir() # 설치된 글꼴을 인식하지 않을 때 캐시메모리를 삭제하고 다시 실행해서 바뀐 설정을 초기화해준다.
```

```python
data = np.random.randint(-100, 100, 50).cumsum() # 누적값을 계산해서 저장
data # 실행할 때마다 랜덤한 결과값이 나온다.
```

```python
plt.plot(range(50),data,'r') #x:range(50), y:data, 'red'
plt.title('시간별 가격 추이')
plt.xlabel('시간')
plt.ylabel('가격') # 깨진다.
```

```python
fm.findSystemFonts(fontext='ttf')[:5] # 해당 운영체제에서 인식하고있는 TTF 폰트 파일이 나온다.
```

```python
matplotlib.matplotlib_fname() # 메모장으로 아래 파일을 열고 세팅을 바꾼다., 설정파일이 어느 경로에 있는지 알 수 있다. #2
```

```python
fm.fontManager.ttflist[:3]
```

```python
[(f.name, f.fname) for f in fm.fontManager.ttflist if 'Nanum' in f.name] # 불러주는 이름, 파일명을 리스트로 저장
# 파일 이름에 'Nanum'이 있는지 확인한다.
```

```python
# 1. font파일이 바로 있는 경우
path = 'NanumBarunpenB.ttf'
fontporp = fm.FontProperties(fname=path,size=17)

plt.plot(range(50),data,'r') #x:range(50), y:data, 'red'
plt.title('시간별 가격 추이', fontproperties=fontporp)
plt.xlabel('시간', fontproperties=fontporp)
plt.ylabel('가격', fontproperties=fontporp)
# 단점은 한글 적는 곳마다 적어야한다.
# 장점은 폰트파일을 직접적으로 같이 줘서 어디 컴퓨터에서 불러와도 깨지진 않는다.
```
