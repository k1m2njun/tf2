```python
def get_yx(query):
    import requests
    headers = {'Authorization': f'KakaoAK {REST_API_KEY}'}
    url = f'https://dapi.kakao.com/v2/local/search/address.json?query={query}'
    r = requests.get(url, headers=headers)
    addr = r.json()
    try: # 오류가 날만한 구간
        return [addr['documents'][0]['y'], addr['documents'][0]['x']]
    except: # 오류가 날 경우 반환값
        return '결과없음'
    
```

```python
# 좌표 나누기
df_dj['위치'] = df_dj['위치'].astype('str')
df_dj['위도'] = df_dj['위치'].str.split(',').str.get(0).str.split('[').str.get(1).str.replace("'","").astype('float')
df_dj['경도'] = df_dj['위치'].str.split(',').str.get(1).str.split(']').str.get(0).str.replace("'","").astype('float')

df_dj
```

```python
loc = [df_dj['위도'].mean(), df_dj['경도'].mean()]
```

```python
import folium
from folium.plugins import MarkerCluster

m = folium.Map(loc, zoom_start=11)
marker_cluster = MarkerCluster().add_to(m)

for i in range(len(df_dj)):
    if df_dj['위치'][i] != '결과없음':
        folium.Marker([df_dj['위도'][i], df_dj['경도'][i]], tooltip=df_dj['설치장소'][i]).add_to(marker_cluster)
        
m
```
