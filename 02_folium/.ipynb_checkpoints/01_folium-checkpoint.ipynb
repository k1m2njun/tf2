라이브러리와 같은 이름의 파일이 있다면 자기 폴더 안에있는 파일을 먼저 갖다 쓴다.py  
라이브러리를 설치할 때 관리자권한 관련 오류가 있다면, 명령프롬프트에서 해봐야한다. 
또 여러군데에 파이썬이 설치되어있으면 경로상 찾지 못할 경우도 있었다.
이럴 때에는 속성에서 경로를 잘 설정해주어야한다.

```python
import folium
```

```python
m = folium.Map(location=[36.316772, 127.4162257], zoom_start=16)
folium.Marker([36.316772, 127.4162257], popup='충남대병원', tooltip='Chung Nam Univ. Hospital').add_to(m)
# tooltip은 마우스를 갖다대면 나타나고, popup은 클릭했을 때 나타난다.
# Shift+Tab
```
