```python
import pandas as pd
```

```python
elet_car = pd.read_csv('서울특별시_자치구별 전기차 용도별 등록현황_20220731.csv', encoding='euc-kr')
elet_car
```

```python
elet_car.plot(x='연월별')
```

```python
elet_car.drop(columns='연료별', inplace=True)
```

```python
elet_car['연도'] = elet_car['연월별'].str.split('-').str.get(0)
elet_car['월'] = elet_car['연월별'].str.split('-').str.get(1)
elet_car['일'] = elet_car['연월별'].str.split('-').str.get(2)
```

```python
elet_car
```

```python
elet_car.to_csv('전처리)구별_전기차_등록_전체.csv', encoding='utf-8-sig')
```
