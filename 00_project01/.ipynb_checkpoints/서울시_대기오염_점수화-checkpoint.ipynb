```python
def jisu(x): # data[['측정항목','평균값','I_LO']]
    if x[0] == 'SO2':
        if x[1] >= 0 and x[1] < 0.021:
            return (50-0)/(0.02-0)*(x[1]-0)+0
        elif x[1] >= 0.021 and x[1] < 0.051:
            return (100-51)/(0.05-0.021)*(x[1]-0.021)+51
        elif x[1] >= 0.051 and x[1] < 0.151:
            return (250-101)/(0.15-0.051)*(x[1]-0.051)+101
        else:
            return (500-251)/(1-0.151)*(x[1]-0.151)+251

    elif x[0] == 'CO':
        if x[1] >= 0 and x[1] < 2.1:
            return (50-0)/(2-0)*(x[1]-0)+0
        elif x[1] >= 2.1 and x[1] < 9.1:
            return (100-51)/(9-2.1)*(x[1]-2.1)+51
        elif x[1] >= 9.1 and x[1] < 15.1:
            return (250-101)/(15-9.1)*(x[1]-9.1)+101
        else:
            return (500-251)/(50-15.1)*(x[1]-15.1)+251

    elif x[0] == 'O3':
        if x[1] >= 0 and x[1] < 0.031:
            return (50-0)/(0.03-0)*(x[1]-0)+0
        elif x[1] >= 0.031 and x[1] < 0.091:
            return (100-51)/(0.09-0.031)*(x[1]-0.031)+51
        elif x[1] >= 0.091 and x[1] < 0.151:
            return (250-101)/(0.15-0.091)*(x[1]-0.091)+101
        else:
            return (500-251)/(0.6-0.151)*(x[1]-0.151)+251

    elif x[0] == 'NO2':
        if x[1] >= 0 and x[1] < 0.031:
            return (50-0)/(0.03-0)*(x[1]-0)+0
        elif x[1] >= 0.031 and x[1] < 0.061:
            return (100-51)/(0.06-0.031)*(x[1]-0.031)+51
        elif x[1] >= 0.061 and x[1] < 0.201:
            return (250-101)/(0.2-0.061)*(x[1]-0.061)+101
        else:
            return (500-251)/(2-0.201)*(x[1]-0.201)+251

    elif x[0] == 'PM10':
        if x[1] >= 0 and x[1] < 31:
            return (50-0)/(30-0)*(x[1]-0)+0
        elif x[1] >= 31 and x[1] < 81:
            return (100-51)/(80-31)*(x[1]-31)+51
        elif x[1] >= 81 and x[1] < 151:
            return (250-101)/(150-81)*(x[1]-81)+101
        else:
            return (500-251)/(600-151)*(x[1]-151)+251

    else:
        if x[1] >= 0 and x[1] < 16:
            return (50-0)/(15-0)*(x[1]-0)+0
        elif x[1] >= 16 and x[1] < 36:
            return (100-51)/(35-16)*(x[1]-16)+51
        elif x[1] >= 36 and x[1] < 76:
            return (250-101)/(75-36)*(x[1]-36)+101
        else:
            return (500-251)/(500-76)*(x[1]-76)+251
        
#https://www.airkorea.or.kr/web/khaiInfo?pMENU_NO=129
```

```python
data['환경지수'] = data[['측정항목','평균값']].apply(lambda x: jisu(x), axis=1)
```

```python
data.to_csv('대기오염환경지수.csv', encoding='utf-8-sig')
```
