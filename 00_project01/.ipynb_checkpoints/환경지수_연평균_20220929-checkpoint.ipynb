```python
# 대기 환경지수 시간별 차트 그리기
```

```python
import pandas as pd
```

```python
data = pd.read_csv('대기오염환경지수.csv')
data.drop(columns='Unnamed: 0', inplace=True)
```

```python
data_seoul = data.groupby(['측정항목','시'])['평균값'].mean()
```

```python
data_seoul = data_seoul.to_frame().reset_index()
```

```python
data_seoul_4 = data_seoul[data_seoul['측정항목']!='PM10']
data_seoul_4 = data_seoul_4[data_seoul_4['측정항목']!='PM2.5']
```

```python
data_seoul_4 = data_seoul_4.reset_index().drop(columns='index') # 4개 물질
```

```python
data_seoul_4
```

```python
data_seoul_2 = data_seoul[data_seoul['측정항목']!='CO']
data_seoul_2 = data_seoul_2[data_seoul_2['측정항목']!='SO2']
data_seoul_2 = data_seoul_2[data_seoul_2['측정항목']!='O3']
data_seoul_2 = data_seoul_2[data_seoul_2['측정항목']!='NO2']
```

```python
data_seoul_2 = data_seoul_2.reset_index().drop(columns='index') # 미세먼지류
```

```python
data_seoul_2
```

# 시각화, 차트

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np  
import seaborn as sns  
sns.set(style="whitegrid", font='NanumGothic')  
sns.lineplot(data=data_seoul_4, x='시', y='평균값', hue='측정항목', palette="seismic_r", linewidth=2, style='측정항목') 
```

```python
sns.set(style="whitegrid", font='NanumGothic')  
sns.lineplot(data=data_seoul_2, x='시', y='평균값', hue='측정항목', palette="seismic_r", linewidth=2, style='측정항목')
```
