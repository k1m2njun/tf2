```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore') 
```

```python
tree_gas = pd.read_csv('ECO100201_산림_가로수_온실가스흡수량_전국.csv', encoding='euc-kr')
tree_gas
```

```python
# ID값 매치를 위한 데이터프레임 새로 만들기
tree_gas_col = tree_gas.columns.to_frame().T
tree_gas_col
```

```python
# ID값 매치를 위한 새 데이터 다시 원 데이터에 취합
tree_gas = tree_gas.append(tree_gas_col)
tree_gas
```

```python
tree_gas = tree_gas.reset_index().drop(columns='index')
tree_gas
```

```python
# 목적, 컬럼 코드를 키값으로 컬럼이름-컬럼코드 merge, 컬럼이름 변경
tree_gas = tree_gas.T
tree_gas
```

# tree_gas

```python
tree_gas_table_info = pd.read_excel('ECO100201_테이블정의서.xlsx', sheet_name='테이블정의서')
tree_gas_table_info
```

```python
tree_gas_table_info = tree_gas_table_info[['Column ID(영문명)_수정본', 'Column Name(한글명)_수정본']]
tree_gas_table_info
```

```python
tree_gas_merge = pd.merge(tree_gas, tree_gas_table_info, left_on=1374, right_on='Column ID(영문명)_수정본', how='outer')
tree_gas_merge
```

```python
tree_gas_merge = tree_gas_merge.T
tree_gas_merge
```

```python
tree_gas_merge = tree_gas_merge.drop([1374])
tree_gas_merge
```

```python
tree_gas_merge = tree_gas_merge.rename(columns=tree_gas_merge.iloc[1375])
tree_gas_merge
```

```python
tree_gas_merge
```

```python
tree_gas_merge.drop([tree_gas_merge.index[1374], tree_gas_merge.index[1375]], inplace=True)
tree_gas_merge
```

```python
tree_gas_merge.info()
```

```python
tree_gas_merge_seoul = tree_gas_merge[tree_gas_merge['광역시명']=='서울특별시']
```

```python
tree_gas_merge_seoul = tree_gas_merge_seoul.reset_index().drop(columns='index')
```

```python
tree_gas_merge_seoul
```

```python
tree_gas_merge_seoul.to_csv('전처리)구별_가로수_온실가스_흡수량_전체.csv', encoding='utf-8-sig')
```

```python
tree_gas_merge_seoul[tree_gas_merge_seoul['기준년도']==2020].to_csv('전처리)구별_가로수_온실가스_흡수량_2020.csv', encoding='utf-8-sig')
```

```python
tree_gas_merge_seoul[tree_gas_merge_seoul['기준년도']==2019].to_csv('전처리)구별_가로수_온실가스_흡수량_2019.csv', encoding='utf-8-sig')
```
