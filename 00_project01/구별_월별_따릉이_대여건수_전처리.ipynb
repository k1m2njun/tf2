```python
REST_API_KEY = '1830579e0e3f456ce3988b002758b9de'
```

```python
def get_yx(query): # 좌표 받기
    import requests
    headers = {'Authorization': f'KakaoAK {REST_API_KEY}'}
    url = f'https://dapi.kakao.com/v2/local/search/address.json?query={query}'
    r = requests.get(url, headers=headers)
    addr = r.json()
    try: # 오류가 날만한 구간
        return [addr['documents'][0]['y'], addr['documents'][0]['x']]
    except: # 오류가 날 경우 반환값
        return '결과없음'
```

```python
ddar_df['위도'] = ddar_df['대여소 그룹'].apply(get_yx).str.get(0)
ddar_df['경도'] = ddar_df['대여소 그룹'].apply(get_yx).str.get(1)
```

# 지도 시각화

```python
ddar_df = pd.read_csv('전처리)구별_월별_따릉이_대여건_2021.csv')
```

```python
ddar_df.drop(columns=['Unnamed: 0', 'index'], inplace=True)
```

```python
ddar_df.월.unique()
```

```python
# 1월
import folium
from folium.plugins import MarkerCluster
ddar_df_loc = [ddar_df['위도'].mean(), ddar_df['경도'].mean()]

m = folium.Map(location=ddar_df_loc, zoom_start=11)
marker_cluster = MarkerCluster().add_to(m)

for i in ddar_df.index:
    if ddar_df['월'][i] == 1:
        folium.Marker([ddar_df['위도'][i], ddar_df['경도'][i]], tooltip=ddar_df['대여 건수'][i]).add_to(marker_cluster)
    
m
```

```python
m = folium.Map(location=ddar_df_loc, zoom_start=11)
marker_cluster = MarkerCluster().add_to(m)

for i in ddar_df.index:
    if ddar_df['월'][i] != '결':
        folium.Marker([ddar_df['위도'][i], ddar_df['경도'][i]], tooltip=ddar_df['대여 건수'][i]).add_to(marker_cluster)
    
m
```
