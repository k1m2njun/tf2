```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('대기오염환경지수.csv')
data.drop(columns='Unnamed: 0', inplace=True)
```

```python
data_group_mon = data.groupby(['측정항목', '월'])['환경지수'].mean()
```

```python
data_group_mon_4 = data_group_mon[data_group_mon['측정항목']!='PM10']
data_group_mon_4 = data_group_mon_4[data_group_mon_4['측정항목']!='PM2.5']
```

```python
data_group_mon_4 # x:월, y:환경지수
```

```python
sns.set_theme(style="ticks", font='NanumGothic', font_scale=0.8)
grid = sns.FacetGrid(data_group_mon_4, col='측정항목', col_wrap=4, palette="tab20c")
grid.refline(y=50, linestyle=":")
grid.map(plt.plot, "월", "환경지수", marker="o")
grid.set(xlim=(0,13), ylim=(0, 60))
grid.fig.tight_layout(h_pad=1)
```

```python
pal = sns.cubehelix_palette(1, rot=-.25, light=.7)
g = sns.FacetGrid(data_group_mon_4, row="측정항목", hue="측정항목", aspect=3, height=3, palette=pal)

# Draw the densities in a few steps

g.map(sns.kdeplot, "월", clip_on=False, color="b", lw=2, bw_adjust=0.1)

# passing color=None to refline() uses the hue mapping
g.refline(y=60, linewidth=2, linestyle="-", color=None, clip_on=False)


# Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, 5, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)


g.map(label, "환경지수")

# Set the subplots to overlap
g.figure.subplots_adjust(hspace=1)

# # Remove axes details that don't play well with overlap
# g.set_titles("")
# g.set(yticks=[], ylabel="")
# g.despine(bottom=True, left=True)
```

```python
sns.set_theme(style="whitegrid", font='NanumGothic')

sns.lineplot(data=data_group_mon_4, x='월', y='환경지수', hue='측정항목', palette="tab10", linewidth=1.5, markers='s')
```

```python
palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
g = sns.relplot(
    data=data_group_mon_4,
    x="월", y="환경지수",
    hue="측정항목",
    kind="line",
    height=5, aspect=3, facet_kws=dict(sharex=False), marker='s'
)

g.refline(y=50, linewidth=1, linestyle=":", color='k', clip_on=False)
```
