```python
df2
```

```python
data3 = pd.read_csv('03_ex/서울시 대기오염 측정항목 정보.csv',encoding='cp949')
```

```python
data3.columns
```

```python
df3 = data3[['측정항목 코드', '측정항목 명(줄임 명칭)','측정단위']]
```

```python
df3
```

```python
#관련 데이터값으로 연결하여 데이터 join 하기
```

```python
merge_inner = pd.merge(df1, df2)
```

```python
df_m= merge_inner
```

```python
merge_inner = pd.merge(df_m, df3)
```

```python
df_m = merge_inner
```

```python
df_m
```

```python
df_m.info()
```

```python
df_m["측정일시"] = df_m["측정일시"].astype('str') # 측정일시를 연도, 월, 일, 시 로 분리 추출
```

```python
df_m['연도'] = df_m["측정일시"].str[:4]
df_m['월'] = df_m["측정일시"].str[4:6]
df_m['일'] = df_m["측정일시"].str[6:8]
df_m['시'] = df_m["측정일시"].str[8:10]
```

```python
df_m.columns
```

```python
df_m = df_m[['측정소 코드', '측정항목 코드', '평균값', '측정기 상태', '측정소 이름', '측정소 주소',
       '측정항목 명(줄임 명칭)', '측정단위', '연도', '월', '일', '시']]
```

```python
df_m.sample(5)
```

```python
# # 정상으로 작동하는 측정기 데이터를 제외하고 모두 삭제
A = df_m[df_m['측정기 상태'] != 0].index
df_m.drop(A,axis='index',inplace=True)
```

```python
# 단위 맞추기 ㎍/㎥ -> ppm
def ex(x):
    if x[0] == '㎍/㎥':
        return x[1]*0.001
    else:
        return x[1]
    
df_m['평균값']=df_m[['측정단위','평균값']].apply(lambda x : ex(x),axis=1)
```

```python
df_m.sample(10)
```

```python
df_m = df_m[['평균값','측정기 상태', '측정소 이름','측정소 주소','측정항목 명(줄임 명칭)','월']]
df_m
```

```python
df_m.info()
```

```python
# 평균값 중 이상치 삭제
B = df_m[df_m['평균값'] == -9999.0].index
df_m.drop(B,axis='index',inplace=True)
```
